Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Upload student-mat.csv in Colab first




```python
df = pd.read_csv('/content/student-mat.csv', sep=';')
df.head()
```

```python
df = df.drop_duplicates()

# Verify new shape
print("Shape after removing duplicates:", df.shape)
```

```python
print(df.duplicated().sum())
```

Analysis Questions
* Average final grade (G3)





```python
avg_grade = df['G3'].mean()
print("Average Final Grade:", avg_grade)
```

Students Scored Above 15





```python
high_score = df[df['G3'] > 15].shape[0]
print("Students scored above 15:", high_score)
```

Study Time Correlation with Performance

```python
correlation = df['studytime'].corr(df['G3'])
print("Correlation between Study Time and G3:", correlation)
```

Which Gender Performs Better?

```python
gender_avg = df.groupby('sex')['G3'].mean()
print(gender_avg)
```

Visualizations
🔹 Histogram of Final Grades

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['G3'], bins=10, kde=True)
plt.title("Distribution of Final Grades")
plt.xlabel("G3 Grade")
plt.ylabel("Count")
plt.show()
```

Scatter Plot: Study Time vs Grades

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

Bar Chart: Male vs Female Average Score

```python
plt.figure(figsize=(8,5))
gender_avg.plot(kind='bar')
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average G3 Score")
plt.show()
```
